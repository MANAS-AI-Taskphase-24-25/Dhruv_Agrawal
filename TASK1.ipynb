```python
np.random.seed(0)
weights = np.random.rand(X_train.shape[1])
bias = 0.0
learning_rate = 0.01
num_iterations = 10000
m = len(y_train)
```

```python
def calculate_metrics(y_actual, y_predicted):
    """
    Calculate key regression metrics between actual and predicted values.

    Parameters:
        y_actual (array-like): True values
        y_predicted (array-like): Predicted values

    Returns:
        tuple: Mean Squared Error (MSE), Root Mean Squared Error (RMSE), and R-squared (RÂ²) score
    """
    # Calculate Mean Squared Error (MSE): Measures the average squared error between actual and predicted values
    mean_squared_error = np.mean((y_actual - y_predicted) ** 2)
    
    # Calculate Root Mean Squared Error (RMSE): Gives the error in the same units as the original output
    root_mean_squared_error = np.sqrt(mean_squared_error)
    
    # Calculate R-squared (RÂ²): Proportion of variance in the actual values explained by the model
    total_variance = np.sum((y_actual - np.mean(y_actual)) ** 2)  # Total sum of squares
    residual_variance = np.sum((y_actual - y_predicted) ** 2)     # Residual sum of squares
    r_squared = 1 - (residual_variance / total_variance)
    
    return mean_squared_error, root_mean_squared_error, r_squared
```

```python
def cost_function(X, y, weights, bias):
    """
    Compute the Mean Squared Error (MSE)-based cost for linear regression.

    Parameters:
        X (array-like): Feature matrix for the training data
        y (array-like): Target values for the training data
        weights (array-like): Model weights for each feature
        bias (float): Bias term for the model

    Returns:
        float: The calculated cost based on Mean Squared Error
    """
    # Total number of training examples
    m = len(y)
    
    # Calculate predictions by applying the linear regression model
    predictions = X.dot(weights) + bias
    
    # Calculate the Mean Squared Error cost
    cost = (1 / (2 * m)) * np.sum((predictions - y) ** 2)
    
    return cost
```

```python
def gradient_descent(X, y, weights, bias, learning_rate, iterations):
    m = len(y)  # Number of train examples
    cost_history = np.zeros(iterations)
    
    for i in range(iterations):
        predictions = X.dot(weights) + bias 
        errors = predictions - y
        
        # Calculate gradients for weights and bias
        weight_gradient = (1 / m) * X.T.dot(errors)
        bias_gradient = (1 / m) * np.sum(errors)
        
        # Update weights and bias
        weights -= learning_rate * weight_gradient
        bias -= learning_rate * bias_gradient
        
        # Calculate cost with separate weights and bias
        cost = cost_function(X, y, weights, bias)
        cost_history[i] = cost

        # Output metrics every 100 iterations
        if i % 100 == 0:
            mse, rmse, r2 = calculate_metrics(y, predictions)
            print(f"Iteration {i}: MSE: {mse:.2f}, RMSE: {rmse:.2f}, RÂ²: {r2:.4f}, Cost: {cost}")
    
    return weights, bias, cost_history
```

```python
def linear_regression(X, y, learning_rate, iterations):
    n = X.shape[1]  # Number of features
    weights = np.zeros(n)  # Initialize weights
    bias = 0.0  # Initialize bias

    weights, bias, cost_history = gradient_descent(X, y, weights, bias, learning_rate, iterations)
    return weights, bias, cost_history
```

```python
learning_rate = 0.05
iterations = 1000

weights, bias, cost_history = linear_regression(X_train, y_train, learning_rate, iterations)

print("\nFinal cost:", cost_history[-1])
print("Final weights:", weights)
print("Final bias:", bias)
```

```python
def predict(X, weights, bias):
    raw_predictions = X.dot(weights) + bias 
    capped_predictions = np.maximum(raw_predictions, 0)  # capping predictions at a minimum of 0 as prices cannot be negative
    return capped_predictions

predictions = predict(X_test, weights, bias)

# Display the first 10 predictions and actual values
print("Predictions on test set:", predictions[:10])
print("Actual values on test set:", y_test[:10])
```
